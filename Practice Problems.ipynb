## Individual Assignment 1: 1b

The Fibonacci sequence 0, 1, 1, 2, 3, 5, 8, ... is defined as follows: F1 = 0, F2 = 1, and Fn =
Fn−1 + Fn−2 for all n ≥ 3. Your program will request a positive integer n, and output the sum
of the first n Fibonacci numbers, the sum of the squares of the first n Fibonacci numbers, as
well as the elapsed time for the computation.

```python
#Requests a positive integer n:
n = int (input("Enter a positive integer n: "))

#Creates a list with all fibonacci numbers up to the nth:
fnumbers = [0,1]
index = 3

lastfib = 1
secondlastfib = 0

fibsum = 1
squaresum = 1

while index <= n:
    
    #Finds the next fibonacci number and adds it to the list
    newfib = lastfib + secondlastfib
    fnumbers.append(newfib)
    
    #Records the new last and second-last fibonacci numbers
    secondlastfib = lastfib
    lastfib = newfib
    
    #Running tally of the sum:
    fibsum = fibsum + newfib
    
    #Running tally of the sum of the squares:
    squaresum = squaresum + (newfib**2)
    
    #Adds 1 to the index
    index += 1
    
#Special case if n = 1
if n == 1:
    fnumbers = [0]
    fibsum = 0
    squaresum = 0
    
#Prints list of n fibbonacci numbers, their sum and the sum of their squares:
print()
print(f"The first {n} fibonacci numbers are:")
print(fnumbers)
print()
print(f"The sum of these numbers is {fibsum}.")
print(f"The sum of their squares is {squaresum}.")
```

## Individual Assignment 2: 2a

Code the BubbleSort algorithm, i.e., the comparison-based process we applied in Adventure
Pursuits to arrange ourselves in the right order. Your Java program will output the number of
pairwise "swaps" needed to sort the string into the correct order.

```python
#Collecting user input:
inputstring = input("Please enter a string of unsorted integers: ")

#Splits the string into a list of numbers based on the white space:
numbers_str = inputstring.split()

#Converts the strings to integers
numbers = list(map(int, numbers_str))

#Counts how many numbers were entered
length = len(numbers)

index = 0
swap_count = 1000
total_swaps = 0

#Performs another check of the list as long as swaps are still being made:
while swap_count > 0:

    #Resets swap count and index:
    swap_count = 0
    index = 0
    
    #Performs one check of all the pairs in the list:
    while index < (length-1):
    
        #If a pair of numbers is out of order, perform a swap:
        if numbers[index] > numbers [(index + 1)]:
        
            #Swap:
            placeholder = numbers[index]
            numbers[index] = numbers[(index + 1)]
            numbers[(index + 1)] = placeholder
            
            #Add one to the swap count:
            swap_count += 1
            total_swaps += 1
            
        #Iterates the index to look at the next pair:
        index += 1
            
            
#Prints the sorted list of numbers:
print()
print("The sorted list is:")
print(numbers)
print()
print(f"Swaps required: {total_swaps}")
```

In class today, we discussed algorithms known as "Monte Carlo Experiments", where we use repeated random sampling to determine an excellent approximation to an exact probability. We solved the Buffon Needle problem, showing that if a 1-unit needle is randomly dropped onto a table ruled with parallel lines 1 unit apart, there is a probability of π/2 that this needle will cross one of the lines.

Consider a square ABCD, with side length 2 units. Two points P and Q are chosen independently, at random, on the sides of this square. Determine the probability that the line segment PQ has distance at most 1.

```python
#Taking user input for n:
n = int(input("Enter n, the number of times you wish to run this experiment: "))

#Importing uniform function, which generates a floating point number between two bounds:
from random import uniform

#Importing randint function, which generates a random integer between two bounds:
from random import randint

index = 1
greater_count = 0
less_count = 0

#Each loop performs one random test:
while index <= n:
    
    #Generates p and q as coordinates along the sides of the square between 0 and 1.
    #p is generated as a coordinate along the top side of the square from left to right.
    #q is generated as a coordinate from top to bottom or left to right depending on the side it is on.
    p = uniform(0,1)
    q = uniform(0,1)
    
    #Generates the side q is on relative to the side p is on.
    #1 is the same side, 2 and 3 are left/right adjacent sides, and 4 is the opposite side.
    sideq = randint(1,4)
    
    #Calculates distance if p and q are on the same side (the top):
    if sideq == 1:
    
        distance = abs(p - q)
        
    #Calculates distance if q is on the adjacent side on the left:
    elif sideq == 2:
    
        distance = ((p**2) + (q**2))**(1/2)
       
    #Calculates distance if q is on the adjacent side on the right:
    elif sideq == 3:
    
        distance = (((1-p)**2) + (q**2))**(1/2) 
        
    #Calculates distance if p and q are on the opposite side:
    elif sideq == 4:
    
        distance = ((1**2) + (abs(p-q)**2))**(1/2)
                    
    else:
        print("ERROR")
    
    #Recording if this test yielded a distance greater or less than/equal to 1:
    if distance > 1:
        
        greater_count += 1
                    
    elif distance <= 1:
        
        less_count += 1
                  
    else:
        print("ERROR")
    
    #Iterates index:
    index += 1
                    
#Calculates percentage of tests that gave a distance less than 1:
probability = less_count/n
                    
print(f"The probability is approximately {probability*100}%.")
```
